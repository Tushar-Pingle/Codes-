```python
import pandas as pd
import numpy as np

Licences_sheet1 = pd.read_excel('C:/Users/12367/Documents/all_liquor_licensed_establishments_in_bc.xlsx', sheet_name='Licences')
Service_Areas_sheet2 = pd.read_excel('C:/Users/12367/Documents/all_liquor_licensed_establishments_in_bc.xlsx', sheet_name='Service Areas')
Hours_of_Service_sheet3 = pd.read_excel('C:/Users/12367/Documents/all_liquor_licensed_establishments_in_bc.xlsx', sheet_name='Hours of Service')
Endorsements_sheet4 = pd.read_excel('C:/Users/12367/Documents/all_liquor_licensed_establishments_in_bc.xlsx', sheet_name='Endorsements')
```

```python
Licences_sheet1.rename(columns={'Establishment Address Postal Code':'postal_code'}, inplace=True)
Service_Areas_sheet2.rename(columns={'Licence Number (Licence) (Licence)':'Licence Number'}, inplace=True)
Hours_of_Service_sheet3.rename(columns={'Licence Number (Licence) (Licence)':'Licence Number'}, inplace=True)
Endorsements_sheet4.rename(columns={'Licence Number (Licence) (Licence)':'Licence Number'}, inplace=True)
```

```python
New_dataset = pd.merge(Licences_sheet1, pd.merge(Service_Areas_sheet2, Endorsements_sheet4, on='Licence Number'), on='Licence Number')
New_dataset.head()
```

```python
## Dropping unwanted Columns
Columns_to_drop = ['Establishment Address Street', 'Establishment Address City', 'Third Party Operator',
                    'Licence Type (Licence) (Licence)_x', 'Establishment (Licence) (Licence)_x', 'Name_x', 'Area Number', 'Area Location', 
                    'LG / IN (Licence) (Licence)_x', 'Licence Type (Licence) (Licence)_y', 'Establishment (Licence) (Licence)_y', 'Name_y',
                      'LG / IN (Licence) (Licence)_y', 'Licence Sub Category Id' ]
New_dataset.drop(Columns_to_drop, axis=1, inplace=True)
New_dataset.head()
```

```python
New_dataset.dropna(inplace=True)
New_dataset.isnull().sum()
```
